***
Normal distribution formula

$$ \large f(x) = \frac{1}{\sigma\sqrt{2\pi}} \Large e^{\frac{-(x-\mu)^{2}}{2\sigma² }}$$
***

```python
import matplotlib.pyplot as plt
import numpy as np
from cmath import pi, e
from math import exp, sqrt
```

```python
def mdist(x, sigma = 1, mu = 1):
    return (1)/(sigma*sqrt(2*pi))*exp(-(x-mu)**2/(2*sigma**2))

mm = [-2,-1,0,1,2]

x = np.arange(-5,5,.01)
y = [[mdist(i, sigma=00.2,mu=m) for i in x] for m in mm]
for f in y:
    plt.plot(x,f)
```

```python
# Obtenemos el directorio y los archivos donde se encuentra la información
import os
import matplotlib.pyplot as plt

cwd   = os.getcwd()
files = os.listdir(cwd+'/ramanspectra')

# Iteramos sobre los archivos para ordenar los datos
# Añadiendo cwd y la carpeta de trabajjo
files = [cwd+'/ramanspectra/'+f for f in files]

# Creando el arreglo de contenidos
conts = []
for file in files:
    with open(file, 'r', encoding = 'unicode_escape') as tempfile:
        conts.append(tempfile.read())

values = conts[0].split('#Acquired')[1].split('\n')[1:-1]
spx=[]
spy=[]
for string_values in values:
    x = float(string_values.split('\t')[0])
    y = float(string_values.split('\t')[1])
    spx.append(x)
    spy.append(y)
    

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 25,
        }

plt.rcParams["figure.figsize"] = (15,10)
plt.plot(spx, spy, color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=5)
plt.title('Dispersión Raman para la muestra a 100W', fontdict=font)
plt.show()
```
