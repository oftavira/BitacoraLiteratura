```python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 44100 # Hz
DURATION = 30 # Seconds

def gen_wave(freq, rate,duration):
    x = np.linspace(0,duration,rate, endpoint = False)
    freqs = x*freq
    y = np.sin((2*np.pi) * freqs)
    return x,y

x,y = gen_wave(5, SAMPLE_RATE,DURATION)
plt.plot(x,y)
```

```python
_, nice_tone = gen_wave(400, SAMPLE_RATE, DURATION)
_, noise_tone = gen_wave(4000, SAMPLE_RATE, DURATION)
noise_tone = noise_tone * 0.3

mixed_tone = nice_tone + noise_tone
```

```python
normalized_tone = np.int16((mixed_tone / mixed_tone.max()) * 32767)

plt.plot(normalized_tone[:300])
plt.show()
```

```python
from scipy.io.wavfile import write

# Remember SAMPLE_RATE = 44100 Hz is our playback rate
write("mysinewave.wav", SAMPLE_RATE, normalized_tone)
```
